```python
#X_train = train.drop("Result", axis=1)
#Y_train = train["Result"]
#X_train.shape, Y_train.shape, test.shape

y = train["Result"]
features = ["O Rk 1", "O Pts/G 1", "O Yds/G 1", "D Rk 1", "D Pts/G 1", "D Yds/G 1", "TO 1", "O Rk 2", "O Pts/G 2", "O Yds/G 2", "D Rk 2", "D Pts/G 2", "D Yds/G 2", "TO 2"]
X = pd.get_dummies(train[features])
X_test = pd.get_dummies(test[features])
```

Here I made the variables I will need to make my prediction. My first attempt is commented out because it didn't work, but that's what I wrote. The first variable is all of the data from my dataset except for the result of the game that data is from. The second variable is just the result of that game. This way, the computer will be able to see how the training data relates to the result. I made a list that had all of the columns I want to use in my dataset.

```python
random_forest = RandomForestClassifier(n_estimators=100)

random_forest.fit(X, y)

Y_pred = random_forest.predict(X_test)
print(Y_pred)
```
